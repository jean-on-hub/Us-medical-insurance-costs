```python
#finding the mode of ages
def find_mode(insurance_data):
    mode_frequency = 0
    mode =""
    for item in insurance_data:
        counter = 0
        for age in insurance_data:
            if item["age"] == age["age"]:
                counter +=1
            if counter > mode_frequency:
                mode_frequency = counter 
                mode = item["age"]
    return "the age with the highest frequency is " + mode +" with " +str(mode_frequency)+ " people"
find_mode(insurance_data)
```
